```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def D(steps):
    x = np.zeros(steps)
    for i in range(1, steps):
        x[i] = x[i - 1] + (np.random.rand() - 0.5) * 2
        
    D = (np.mean(x**2) - np.mean(x)**2) / (steps * 2)
    return D
```

```python
steps = 10000
d_count = 100

appr_D = 0
for i in range(0, 100):
    appr_D += D(10000)

appr_D / 100
```

```python
d_count = 1000
Ds = np.zeros(d_count)

for i in range(0, d_count):        
    Ds[i] = D(steps)
```

```python
plt.figure(1, (10,6))
plt.hist(Ds, bins = 50)
plt.xlabel("D")
plt.ylabel("Count")
plt.show()
```

```python
np.mean(Ds)
```

```python
np.std(Ds)
```

```python
np.std(Ds) / np.sqrt(d_count)
```
